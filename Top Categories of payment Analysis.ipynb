```python
import json 

import numpy as np

from plotly.offline import plot

import matplotlib.pyplot as plt 

import plotly.express as px

import plotly.graph_objs as go


with open("combined_Trans_2018.json", 'r') as f: 

    data = json.load(f) 

    

transaction_data = data['data']['transactionData']


transaction_count = {}


for transaction in transaction_data: 
    
    name = transaction['name']
    
    count  =  transaction['paymentInstruments'][0]['count']
    
    
    if name in transaction_count:
        
        transaction_count[name]  +=  count
        
    else:
        
        transaction_count[name]  =  count

    
    
    
labels   =  list(transaction_count.keys())

values = list(transaction_count.values())


c=["Orange","Black","Orange","Black","Orange"]

plt.bar(labels,values,0.6,color=c)

plt.xticks(np.arange(0,5,step=1),size="x-small",rotation=90)


plt.title("Top Categories of Payments in 2018")
plt.xlabel("Payment_Instruments")
plt.ylabel('Payment_instruments_usage in 2018') 

plt.figure(figsize=(5,5)) 
plt.show()




import json 

import numpy as np



import matplotlib.pyplot as plt 

import plotly.express as px

with open("combined_Trans_2019.json", 'r') as f: 

    data = json.load(f) 

    
transaction_data = data['data']['transactionData']


transaction_count = {}


for transaction in transaction_data: 
    
    name = transaction['name']
    
    count  =  transaction['paymentInstruments'][0]['count']
    
    
    if name in transaction_count:
        
        transaction_count[name]  +=  count
        
    else:
        
        transaction_count[name]  =  count

    
labels   =  list(transaction_count.keys())

values = list(transaction_count.values())


c=["Orange","Black","Orange","Black","Orange"]

plt.bar(labels,values,0.6,color=c)

plt.xticks(np.arange(0,5,step=1),size="x-small",rotation=90)


plt.title("Top Categories of Payments in 2019")
plt.xlabel("Payment_Instruments")
plt.ylabel('Payment_instruments_usage in 2019') 

plt.figure(figsize=(5,5)) 
plt.show()

import json 

import numpy as np



import matplotlib.pyplot as plt 

import plotly.express as px

with open("combined_Trans_2020.json", 'r') as f: 

    data = json.load(f) 

    
transaction_data = data['data']['transactionData']


transaction_count = {}


for transaction in transaction_data: 
    
    name = transaction['name']
    
    count  =  transaction['paymentInstruments'][0]['count']
    
    
    if name in transaction_count:
        
        transaction_count[name]  +=  count
        
    else:
        
        transaction_count[name]  =  count

    
labels   =  list(transaction_count.keys())

values = list(transaction_count.values())


c=["Orange","Black","Orange","Black","Orange"]

plt.bar(labels,values,0.6,color=c)

plt.xticks(np.arange(0,5,step=1),size="x-small",rotation=90)


plt.title("Top Categories of Payments in 2020")
plt.xlabel("Payment_Instruments")
plt.ylabel('Payment_instruments_usage in 2020') 

plt.figure(figsize=(5,5)) 
plt.show() 


import json 

import numpy as np



import matplotlib.pyplot as plt 

import plotly.express as px

with open("combined_Trans_2021.json", 'r') as f: 

    data = json.load(f) 

    
transaction_data = data['data']['transactionData']


transaction_count = {}


for transaction in transaction_data: 
    
    name = transaction['name']
    
    count  =  transaction['paymentInstruments'][0]['count']
    
    
    if name in transaction_count:
        
        transaction_count[name]  +=  count
        
    else:
        
        transaction_count[name]  =  count

    
labels   =  list(transaction_count.keys())

values = list(transaction_count.values())


c=["Orange","Black","Orange","Black","Orange"]

plt.bar(labels,values,0.6,color=c)

plt.xticks(np.arange(0,5,step=1),size="x-small",rotation=90)


plt.title("Top Categories of Payments in 2021")
plt.xlabel("Payment_Instruments")
plt.ylabel('Payment_instruments_usage in 2021') 

plt.figure(figsize=(5,5)) 
plt.show()


import json 

import numpy as np



import matplotlib.pyplot as plt 

import plotly.express as px

with open("combined_Trans_2022.json", 'r') as f: 

    data = json.load(f) 

    
transaction_data = data['data']['transactionData']


transaction_count = {}


for transaction in transaction_data: 
    
    name = transaction['name']
    
    count  =  transaction['paymentInstruments'][0]['count']
    
    
    if name in transaction_count:
        
        transaction_count[name]  +=  count
        
    else:
        
        transaction_count[name]  =  count

    
labels   =  list(transaction_count.keys())

values = list(transaction_count.values())


c=["Orange","Black","Orange","Black","Orange"]

plt.bar(labels,values,0.6,color=c)

plt.xticks(np.arange(0,5,step=1),size="x-small",rotation=90)


plt.title("Top Categories of Payments in 2022")
plt.xlabel("Payment_Instruments")
plt.ylabel('Payment_instruments_usage in 2022') 

plt.figure(figsize=(5,5)) 
plt.show() 
```
