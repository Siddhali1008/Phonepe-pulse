```python
import json
import matplotlib.pyplot as plt

with open("combined_data2018_top5.json", 'r') as f:
 data = json.load(f)
states = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names = [state['name'].upper() for state in states]
registered_users = [state['registeredUsers'] for state in states]


with open("combined_data2019_top5.json", 'r') as f:
 data = json.load(f)
states1 = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names1 = [state['name'].upper() for state in states1]
registered_users1 = [state['registeredUsers'] for state in states1]
 
    
with open("combined_data2020_top5.json", 'r') as f:
 data = json.load(f)
states2 = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names2 = [state['name'].upper() for state in states2]
registered_users2 = [state['registeredUsers'] for state in states2]


with open("combined_data2021_top5.json", 'r') as f:
 data = json.load(f)
states3 = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names3 = [state['name'].upper() for state in states3]
registered_users3 = [state['registeredUsers'] for state in states3]

with open("combined_data2022_top5.json", 'r') as f:
 data = json.load(f)
states4 = sorted(data['data']['states'], key=lambda x: x['registeredUsers'],reverse=True)
state_names4 = [state['name'].upper() for state in states4]
registered_users4 = [state['registeredUsers'] for state in states4]

plt.figure(figsize=(20,10))
plt.plot(state_names, registered_users,label="2018")
plt.plot(state_names, registered_users1,label="2019")
plt.plot(state_names, registered_users2,label="2020")
plt.plot(state_names, registered_users3,label="2021")
plt.plot(state_names, registered_users4,label="2022")

plt.title('Top 5 States for Registered Users in 2018-2022')
plt.xlabel('State'.upper())
plt.ylabel('Number of Registered Users')
plt.legend(edgecolor="black")
plt.show()
```
