```python
import json
import matplotlib.pyplot as plt

with open('C:/Users/shreedhar_gehlot/Desktop/final/phonePe_data/data/aggregated/transaction/country/india/2018/1.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']


transaction_count = {}

for transaction in transaction_data:
    name = transaction['name']
    count = transaction['paymentInstruments'][0]['count']
    if name in transaction_count:
        transaction_count[name] += count
    else:
        transaction_count[name] = count

labels = list(transaction_count.keys())
values = list(transaction_count.values())
plt.figure(figsize=(20,10))
plt.bar(labels, values)
plt.title('Transaction in 1st Quarter of 2018 ')
plt.xlabel('name'.upper())
plt.ylabel('Number of Transaction * 10^7')
plt.show()


with open('C:/Users/shreedhar_gehlot/Desktop/final/phonePe_data/data/aggregated/transaction/country/india/2018/4.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']


transaction_count = {}

for transaction in transaction_data:
    name = transaction['name']
    count = transaction['paymentInstruments'][0]['count']
    if name in transaction_count:
        transaction_count[name] += count
    else:
        transaction_count[name] = count

labels = list(transaction_count.keys())
values = list(transaction_count.values())
plt.figure(figsize=(20,10))
plt.bar(labels, values,color='red')
plt.title('Transaction in 4th Quarter of 2018')
plt.xlabel('name'.upper())
plt.ylabel('Number of Transaction in  * 10^7 ')
plt.show()
```
