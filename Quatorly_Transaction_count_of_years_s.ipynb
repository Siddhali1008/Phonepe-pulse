```python
#To plot the pie chart for types of payment via Phonepe
#Total Payment via PhonePe in the year 2018 Quartorly

import json
import matplotlib.pyplot as plt

with open('2018_combined_data.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']

transaction_count=0
lists=[]

for x in transaction_data:
    
    Count=x['paymentInstruments'][0]['count']
    lists.append(Count)
add1=sum(lists[0:5])
add2=sum(lists[5:10])
add3=sum(lists[10:15])
add4=sum(lists[15:19])
sum_of=list[add1,add2,add3,add4]

labels = ['1st Quarter','2nd Quarter','3rd Quarter','4th Quarter']
values = [add1,add2,add3,add4]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Total Payment via PhonePe in the year 2018 Quartorly')
plt.figure(figsize=(15,15))
plt.show()
print("Total payments in 1st Quarter of 2018-",add1)
print("Total payments in 2nd Quarter of 2018-",add2)
print("Total payments in 3rd Quarter of 2018-",add3)
print("Total payments in 4th Quarter of 2018-",add4)


#Total Payment via PhonePe in the year 2019 Quartorly
import json
import matplotlib.pyplot as plt

with open('2019_combined_data.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']

transaction_count=0
lists=[]

for x in transaction_data:
    
    Count=x['paymentInstruments'][0]['count']
    lists.append(Count)
add1=sum(lists[0:5])
add2=sum(lists[5:10])
add3=sum(lists[10:15])
add4=sum(lists[15:19])
sum_of=list[add1,add2,add3,add4]



labels = ['1st Quarter','2nd Quarter','3rd Quarter','4th Quarter']
values = [add1,add2,add3,add4]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Total Payment via PhonePe in the year 2019 Quartorly')
plt.figure(figsize=(10,10))
plt.show()
print("Total payments in 1st Quarter of 2019-",add1)
print("Total payments in 2nd Quarter of 2019-",add2)
print("Total payments in 3rd Quarter of 2019-",add3)
print("Total payments in 4th Quarter of 2019-",add4)

#Total Payment via PhonePe in the year 2020 Quartorly
import json
import matplotlib.pyplot as plt

with open('2020_combined_data.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']

transaction_count=0
lists=[]

for x in transaction_data:
    
    Count=x['paymentInstruments'][0]['count']
    lists.append(Count)
add1=sum(lists[0:5])
add2=sum(lists[5:10])
add3=sum(lists[10:15])
add4=sum(lists[15:19])
sum_of=list[add1,add2,add3,add4]



labels = ['1st Quarter','2nd Quarter','3rd Quarter','4th Quarter']
values = [add1,add2,add3,add4]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Total Payment via PhonePe in the year 2020 Quartorly')
plt.figure(figsize=(10,10))
plt.show()
print("Total payments in 1st Quarter of 2020-",add1)
print("Total payments in 2nd Quarter of 2020-",add2)
print("Total payments in 3rd Quarter of 2020-",add3)
print("Total payments in 4th Quarter of 2020-",add4)

#Total Payment via PhonePe in the year 2021 Quartorly
import json
import matplotlib.pyplot as plt

with open('2021_combined_data.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']

transaction_count=0
lists=[]

for x in transaction_data:
    
    Count=x['paymentInstruments'][0]['count']
    lists.append(Count)
add1=sum(lists[0:5])
add2=sum(lists[5:10])
add3=sum(lists[10:15])
add4=sum(lists[15:19])
sum_of=list[add1,add2,add3,add4]



labels = ['1st Quarter','2nd Quarter','3rd Quarter','4th Quarter']
values = [add1,add2,add3,add4]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Total Payment via PhonePe in the year 2021 Quartorly')
plt.figure(figsize=(10,10))
plt.show()
print("Total payments in 1st Quarter of 2021-",add1)
print("Total payments in 2nd Quarter of 2021-",add2)
print("Total payments in 3rd Quarter of 2021-",add3)
print("Total payments in 4th Quarter of 2021-",add4)


#Total Payment via PhonePe in the year 2022 Quartorly
import json
import matplotlib.pyplot as plt

with open('2022_combined_data.json') as f:
    data = json.load(f)

transaction_data = data['data']['transactionData']

transaction_count=0
lists=[]

for x in transaction_data:
    
    Count=x['paymentInstruments'][0]['count']
    lists.append(Count)
add1=sum(lists[0:5])
add2=sum(lists[5:10])
add3=sum(lists[10:15])
add4=sum(lists[15:19])
sum_of=list[add1,add2,add3,add4]



labels = ['1st Quarter','2nd Quarter','3rd Quarter','4th Quarter']
values = [add1,add2,add3,add4]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Total Payment via PhonePe in the year 2022 Quartorly')
plt.figure(figsize=(10,10))
plt.show()
print("Total payments in 1st Quarter of 2022-",add1)
print("Total payments in 2nd Quarter of 2022-",add2)
print("Total payments in 3rd Quarter of 2022-",add3)
print("Total payments in 4th Quarter of 2022-",add4)
```
